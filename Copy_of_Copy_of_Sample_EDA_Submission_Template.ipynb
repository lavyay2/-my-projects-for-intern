## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Convert timestamp columns to datetime objects
df['Request timestamp'] = pd.to_datetime(df['Request timestamp'], errors='coerce')
df['Drop timestamp'] = pd.to_datetime(df['Drop timestamp'], errors='coerce')

# Extract time-based features
df['Request hour'] = df['Request timestamp'].dt.hour
df['Request Day of Week'] = df['Request timestamp'].dt.day_name()
df['Request Month'] = df['Request timestamp'].dt.month
df['Request Year'] = df['Request timestamp'].dt.year

# Calculate trip duration for completed trips
df['Trip Duration'] = (df['Drop timestamp'] - df['Request timestamp']).dt.total_seconds() / 60

# Display the first few rows with new columns
display(df.head())
````

### What all manipulations have you done and insights you found?

Answer    I converted the timestamp columns to datetime objects and extracted time features like hour, day of the week, month, and year. I also calculated the trip duration for completed rides.
These steps allow us to analyze:

**When people** request Ubers (peak hours, days, etc.).

**How long** completed trips take.

This information is key to understanding the supply and demand patterns.

#### Chart - 4

````python
# Chart - 4 visualization code
# Create time slots
def get_time_slot(hour):
    if 4 <= hour < 10:
        return 'Morning'
    elif 10 <= hour < 16:
        return 'Daytime'
    elif 16 <= hour < 22:
        return 'Evening'
    else:
        return 'Night'

df['Time Slot'] = df['Request hour'].apply(get_time_slot)

# Group by pickup point and time slot, then count status
status_by_pickup_timeslot = df.groupby(['Pickup point', 'Time Slot'])['Status'].value_counts().unstack().fillna(0)

status_by_pickup_timeslot.plot(kind='bar', stacked=True, figsize=(14, 7))
plt.title('Uber Request Status Distribution by Pickup Point and Time Slot')
plt.xlabel('Pickup Point and Time Slot')
plt.ylabel('Number of Requests')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Status')
plt.tight_layout()
plt.show()
````

Answer  I used a stacked bar chart to show how request statuses are distributed by both pickup point and time of day, making it easy to compare the breakdown across different locations and times.

##### 2. What is/are the insight(s) found from the chart?

Answer Looking at the chart showing Uber request status distribution by pickup point and time slot, here are some insights:

Morning Rush (City to Airport): There's a significant number of cancelled requests from the City during the Morning time slot (4 AM to 10 AM).

This suggests that many people requesting rides from the City to the Airport in the morning are unable to get one, likely due to a shortage of drivers in the City willing to go to the Airport during that busy time.

Based on the exploratory data analysis of the Uber request data, the following key conclusions can be drawn regarding the supply-demand gap:

1.  **Significant Unfulfilled Demand:** A substantial portion of Uber requests remain unfulfilled, primarily due to cancellations and the lack of available cars. This highlights a significant gap between the demand for Uber rides and the available supply.
2.  **Peak Hour Imbalances:** The supply-demand gap is particularly severe during specific peak hours. The morning rush hour (roughly 4 AM to 10 AM) in the City experiences a high rate of cancellations, suggesting drivers might be reluctant to accept certain trips during this busy period, possibly due to traffic or destination. The evening and night hours (roughly 4 PM to 4 AM) at the Airport suffer from a critical shortage of available cars, indicating insufficient driver supply to meet the high demand from arriving passengers.
3.  **Location-Specific Problems:** The nature of the supply-demand gap differs between pickup locations. The City's main issue during its peak is cancellations, while the Airport's main issue during its peak is the lack of available vehicles.
4.  **Limited Data Scope:** The analysis was limited by the availability of data for only two specific dates (November 7th and December 7th, 2016) and only two days of the week (Monday and Wednesday). A more comprehensive dataset spanning a longer period and covering all days of the week would be beneficial for identifying broader trends and seasonal patterns.

**Recommendations to address the gap:**

*   Implement targeted incentives to encourage drivers to be available in the City during morning peak hours and at the Airport during evening/night peak hours.
*   Investigate the specific reasons for driver cancellations in the City during the morning and address those factors (e.g., improve routing algorithms, manage rider expectations on wait times for certain destinations).
*   Optimize driver positioning and allocation strategies, particularly at the Airport during evening and night, potentially using predictive analytics to anticipate demand.
*   Improve communication with both drivers and riders regarding expected demand, surge pricing, and potential wait times during peak periods.

Addressing these specific issues at the identified critical times and locations is crucial for reducing the supply-demand gap, improving operational efficiency, and enhancing customer satisfaction for Uber's services, especially for airport transportation.
